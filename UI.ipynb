```python
from pydub import AudioSegment
import math
import os

def split_audio(input_file, output_directory,model_name):
    segment_duration=3
    # Load the audio file
    audio = AudioSegment.from_file(input_file)
    # Calculate the total number of segments
    total_segments = math.ceil(len(audio) / (segment_duration * 1000))

    # Create the output directory if it doesn't exist
    os.makedirs(output_directory, exist_ok=True)

    # Split the audio into segments
    for i in range(total_segments):
        start_time = i * segment_duration * 1000
        end_time = min((i + 1) * segment_duration * 1000, len(audio))
        segment = audio[start_time:end_time]

        # Save each segment to the output directory
        output_file = os.path.join(output_directory, f'{model_name}_{i + 1}.wav')
        segment.export(output_file, format='wav')


```

```python
import shutil
import os
import glob

def copy_file(source_path, destination_path, new_filename):
    # Create the full path to the source file using glob
    matching_files = glob.glob(source_path, recursive=True)

    if matching_files:
        # If there are matching files, use the first one found
        source_file = matching_files[0]
        print(f"Copying file from {source_file} to {destination_path}/{new_filename}")

        # Combine the destination directory and the new filename to get the full destination path
        destination_file = os.path.join(destination_path, new_filename)

        # Use shutil.copy to copy the file
        shutil.copy(source_file, destination_file)

        print("File copied successfully.")
    else:
        print(f"No matching files found for {source_path}")

# Example usage:
source_directory = r'C:\Users\muham\Downloads'
filename_pattern = 'audio.mp3'

destination_directory = r'C:\Users\muham\Documents\NewDestination'
new_filename = 'new_audio.mp3'

# Call the copy_file function
copy_file(os.path.join(source_directory, '**', filename_pattern), destination_directory, new_filename)
```
